```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
```

```python
def getDays():
    days = []
    for _ in range(1,32):
        if _ < 10:
            days.append(f"2021-01-0{_}")
        else:
            days.append(f"2021-01-{_}")
    for _ in range(1,29):
        if _ < 10:
            days.append(f"2021-02-0{_}")
        else:
            days.append(f"2021-02-{_}")
    for _ in range(1,32):
        if _ < 10:
            days.append(f"2021-03-0{_}")
        else:
            days.append(f"2021-03-{_}")
    for _ in range(1,31):
        if _ < 10:
            days.append(f"2021-04-0{_}")
        else:
            days.append(f"2021-04-{_}")
    for _ in range(1,32):
        if _ < 10:
            days.append(f"2021-05-0{_}")
        else:
            days.append(f"2021-05-{_}")
    for _ in range(1,31):
        if _ < 10:
            days.append(f"2021-06-0{_}")
        else:
            days.append(f"2021-06-{_}")
    for _ in range(1,32):
        if _ < 10:
            days.append(f"2021-07-0{_}")
        else:
            days.append(f"2021-07-{_}")
    for _ in range(1,32):
        if _ < 10:
            days.append(f"2021-08-0{_}")
        else:
            days.append(f"2021-08-{_}")
    for _ in range(1,31):
        if _ < 10:
            days.append(f"2021-09-0{_}")
        else:
            days.append(f"2021-09-{_}")
    for _ in range(1,32):
        if _ < 10:
            days.append(f"2021-10-0{_}")
        else:
            days.append(f"2021-10-{_}")
    for _ in range(1,31):
        if _ < 10:
            days.append(f"2021-11-0{_}")
        else:
            days.append(f"2021-11-{_}")
    for _ in range(1,32):
        if _ < 10:
            days.append(f"2021-12-0{_}")
        else:
            days.append(f"2021-12-{_}")
    return days
```

```python
columns = [
'Latitude',
'Longitude',
'acq_date',
'Temperatura máxima (ºC)',
'Temperatura mínima (ºC)',
'Temperatura media (ºC)',
'Racha (km/h)',
'Velocidad máxima (km/h)',
'Precipitación 00-24h (mm)',
'Precipitación 00-06h (mm)',
'Precipitación 06-12h (mm)',
'Precipitación 12-18h (mm)',
'Precipitación 18-24h (mm)']

datasetRandomPoints = pd.DataFrame(columns=columns)

days = getDays()

for day in days:
    for _ in range(10):
        lat = random.uniform(37, 43)
        lon = random.uniform(-6.5, 0)
        datasetRandomPoints = datasetRandomPoints.append({'Latitude': lat, 'Longitude': lon, 'acq_date': day}, ignore_index=True)

datasetRandomPoints
for index in range(len(datasetRandomPoints)):
    day = datasetRandomPoints.iloc[_, 2]
    for colum in columns[3:]:
        df = pd.read_csv(f'./ClimaDataSmall/Aemet{day}.csv')
        columnIndex = df.columns.get_loc(colum)
        df = df.iloc[:,[-1,-2,columnIndex]]
        df = df.query('Latitude > 30 and Latitude < 44 and Longitude > -20')
        if str(df.iloc[1,2]).find(' ') != -1:
            for _ in range(len(df)):
                df.iloc[_,2] = float(df.iloc[_,2][:df.iloc[_,2].find(' ')])

        df2 = df.drop([colum], 1)
        df3 = df[colum]
        lat = df2['Latitude']
        lon = df2['Longitude']
        values = np.array(df3,dtype='float64')
        arraylat = np.array(lat,dtype='float64')
        arraylon = np.array(lon,dtype='float64')
        est_u = griddata((arraylat, arraylon), values, (datasetRandomPoints.iloc[index,0],datasetRandomPoints.iloc[index,1]), method='linear')
        columnIndex = datasetRandomPoints.columns.get_loc(colum)
        datasetRandomPoints.iloc[index,columnIndex] = est_u


datasetRandomPoints.to_csv('datasetRandomPoints.csv', index=False)
```
